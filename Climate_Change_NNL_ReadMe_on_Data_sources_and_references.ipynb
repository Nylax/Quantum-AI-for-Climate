# Introduction
This project focuses on analyzing data from multiple sources to understand the South Atlantic Anomaly (SAA) and its potential impact on climate change. The SAA is a region where the Earth's magnetic field is significantly weaker, affecting satellite operations and possibly influencing climate patterns. We utilize data from the SWARM satellite constellation, the Copernicus satellite, and the Argentinian Meteorological Agency to conduct this analysis.
